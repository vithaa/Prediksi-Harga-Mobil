Jumlah Mobil Berdasarkan model

```python
models = df.groupby('model').count()[['tax']].sort_values(by='tax',ascending=True).reset_index()
models = models.rename(columns={'tax':'NumberOfCars'})
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['model'], y=models['NumberOfCars'], color='red')
plt.xticks(rotation=60)
```

Ukuran Mesin

```python
engine = df.groupby('engineSize').count()[['tax']].sort_values(by='tax').reset_index()
engine = engine.rename(columns={'tax':'count'})
fig = plt.figure(figsize=(15,5))
sns.barplot(x=engine['engineSize'], y=engine['count'], color='red')
plt.xticks(rotation=60)
```

Seleksi Fitur

```python
fitur = ['year','mileage','tax','mpg','engineSize']
x = df[fitur]
y = df['price']
x.shape, y.shape
x
```

Split Data Training dan Data Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 70)
y_test.shape,y_train.shape
```

Membuat Model 

```python
# model Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred1 = lr.predict(x_test)
score1 = lr.score(x_test, y_test)
print('Akurasi Model Regresi Linear = ',score1)

import matplotlib.pyplot as plt
plt.scatter(y_test, pred1)
plt.plot(y_test, y_test, color='red', linestyle='--')
plt.xlabel('Data Point')
plt.ylabel('Score')
plt.title('Model Regresi Linear - Akurasi')
plt.show()
```

Membuat Inputan Model Regresi Linear

```python
# Format Input (Year, mileage, tax, mpg, engineSize)
input_data = np.array([[2019,5000,145,30.2,2]])
prediction = lr.predict(input_data)
print('Estimasi Harga Mobil Dalam EUR : ', prediction)
```
